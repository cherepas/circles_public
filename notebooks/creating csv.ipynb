```python
minv = np.array([100, 100, 100])
maxv = np.array([0, 0, 0])
for i in range(len(cip)):
    pcd = open3d.io.read_point_cloud(cip[i])
    img = np.asarray(pcd.points)
    img -= np.mean(img,0)
    minc = np.min(img, axis=0)
    maxc = np.max(img, axis=0)
    for j in range(3):
        if minc[j] < minv[j]:
            minv[j] = minc[j]
        if maxc[j] > maxv[j]:
            maxv[j] = maxc[j]
```
